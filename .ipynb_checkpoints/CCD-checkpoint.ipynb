```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
u_MAG = []
g_MAG = []
i_MAG = []
z_MAG = []
ug_MAG = []
gi_MAG = []
i_classstar = []
with open('1261_full.txt','r') as data:
    for line in data:
        linevals = line.split("\t")
        i_classstar.append(float(linevals[11]))
        u_MAG.append(float(linevals[3]))
        g_MAG.append(float(linevals[6]))
        i_MAG.append(float(linevals[9]))
        z_MAG.append(float(linevals[12]))
        ug_MAG.append(float(linevals[3]) - float(linevals[6]))
        gi_MAG.append(float(linevals[6]) - float(linevals[9]))
    data.close()
ug_MAG = np.array(ug_MAG)
gi_MAG = np.array(gi_MAG)
keep = (np.array(g_MAG) < 24) & (np.array(i_classstar) > 0.8)
keep2 = (np.array(g_MAG) < 24) & (np.array(i_classstar) < 0.2)
plt.scatter(ug_MAG[keep], gi_MAG[keep],s = 5, c = 'red')
plt.scatter(ug_MAG[keep2], gi_MAG[keep2],s = 15, c = 'blue')
plt.xlim(-1,4)
plt.ylim(-1,3)
plt.ylabel("g-i")
plt.xlabel("u-g")
plt.show()
```
